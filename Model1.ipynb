```python
import csv
import json

import requests


def replace_short_forms(sentence):
    short_forms = {}
    with open('accronym.csv', newline='') as csvfile:
        reader = csv.reader(csvfile, delimiter=',', quotechar='"')
        for row in reader:
            short_forms[row[0]] = row[1]
    
    words = sentence.split()
    for i, word in enumerate(words):
        if word.lower() in short_forms:
            words[i] = short_forms[word.lower()]
    
    return ' '.join(words)

def spell_check(sentence, whitelist):
    api_key = "bd280f0248e942a3932c90bb79da44ef"
    endpoint = "https://api.bing.microsoft.com/v7.0/SpellCheck"
    headers = {
        "Content-Type": "application/x-www-form-urlencoded",
        "Ocp-Apim-Subscription-Key": api_key,
    }
    data = {
        "text": sentence,
        "mode": "proof",
        "mkt": "en-US",
    }
    response = requests.post(endpoint, headers=headers, data=data)
    result = json.loads(response.content.decode("utf-8"))

    flagged_tokens = result.get("flaggedTokens", [])
    corrected_sentence = sentence

    for token in flagged_tokens:
        if token["token"] not in whitelist:
            corrected_sentence = corrected_sentence.replace(token["token"], token["suggestions"][0]["suggestion"])
    
    return corrected_sentence

def preprocess(text):
    text = text.lower()
    
    text = replace_short_forms(text)
    
    text = emoji.demojize(text)
    text = text.replace(":", "")
    
    text = re.sub(r'(\w)([A-Z])', r'\1 \2', text)
    
    # Remove unnecessary white space
    text = re.sub(' +', ' ', text)
    
    # Correct misspelling and add space where needed
    whitelist = ['copay', 'srf', 'mir', 'cordastin', 'yo', 'ahoy', 'aloha', 'sup', 'yoohoo', 'hey hey', 'hiya', 'hola', 'nice', 'sample', 'samples']
    text = str(spell_check(text, whitelist))
    
    # Remove special characters and punctuation
    text = re.sub(r'[^\w\s]', '', text)
    
    # Remove leading/trailing white space
    text = text.strip()
    
    return text
```

```python
import torch
import numpy as np
from transformers import AutoTokenizer, AutoModel

# Mean pooling to obtain sentence embeddings
def mean_pooling(model_output, attention_mask):
    token_embeddings = model_output.last_hidden_state
    input_mask_expanded = attention_mask.unsqueeze(-1).expand(token_embeddings.size()).float()
    return torch.sum(token_embeddings * input_mask_expanded, 1) / torch.clamp(input_mask_expanded.sum(1, keepdim=True), min=1e-9)

tokenizer = AutoTokenizer.from_pretrained('distilroberta-base')  
model = AutoModel.from_pretrained('distilroberta-base') 

sentence_embeddings = []

for sentence in train_utterances:
    encoded_input = tokenizer(sentence, padding=True, truncation=True, return_tensors='pt')
    with torch.no_grad():
        model_output = model(**encoded_input)
    embeddings = mean_pooling(model_output, encoded_input['attention_mask'])
    sentence_embeddings.append(embeddings)

sentence_embeddings = torch.stack(sentence_embeddings)

print("Sentence Embeddings:")
print(sentence_embeddings.shape)
```

```python
import torch
from transformers import RobertaTokenizer, RobertaModel
from sklearn.ensemble import RandomForestClassifier
from sklearn.preprocessing import LabelEncoder
import re

def preprocess(sentence):
    sentence = sentence.lower()
    sentence = re.sub(r'(\w)([A-Z])', r'\1 \2', sentence)
    sentence = re.sub(' +', ' ', sentence)
    sentence = sentence.strip()
    return sentence

def get_roberta_embeddings(sentences, model, tokenizer):
    embeddings = []
    for sentence in sentences:
        sentence = preprocess(sentence)
        inputs = tokenizer(sentence, return_tensors='pt', padding=True, truncation=True)
        with torch.no_grad():
            outputs = model(**inputs)
        embeddings.append(outputs.last_hidden_state.mean(dim=1).numpy()[0])
    return embeddings

def train_rf_classifier(embeddings, labels):
    label_encoder = LabelEncoder()
    encoded_labels = label_encoder.fit_transform(labels)
    rf_classifier = RandomForestClassifier(n_estimators=100, random_state=42)
    rf_classifier.fit(embeddings, encoded_labels)
    return rf_classifier, label_encoder

def predict_sentiment(sentence, model, tokenizer, rf_classifier, label_encoder):
    test_sentence = preprocess(sentence)
    test_inputs = tokenizer(test_sentence, return_tensors='pt', padding=True, truncation=True)
    with torch.no_grad():
        test_outputs = model(**test_inputs)
    test_embedding = test_outputs.last_hidden_state.mean(dim=1).numpy()[0]
    predicted_label = rf_classifier.predict([test_embedding])[0]
    predicted_sentiment = label_encoder.inverse_transform([predicted_label])[0]
    predicted_probabilities = rf_classifier.predict_proba([test_embedding])[0]
    predicted_sentiment_score = predicted_probabilities[predicted_label]
    return predicted_sentiment, predicted_sentiment_score

if __name__ == "__main__":
    tokenizer = AutoTokenizer.from_pretrained('distilroberta-base')  
    model = AutoModel.from_pretrained('distilroberta-base') 

    sentences = ["This is a great product.", "I am not satisfied with this.", "It's okay, not too bad."]
    labels = ["Positive", "Negative", "Neutral"]

    embeddings = get_roberta_embeddings(sentences, model, tokenizer)
    rf_classifier, label_encoder = train_rf_classifier(embeddings, labels)

    test_sentence = "$ESI on lows, down $1.50 to $2.50 BK a real possibility"
    predicted_sentiment, predicted_sentiment_score = predict_sentiment(test_sentence, model, tokenizer, rf_classifier, label_encoder)

    print("Input Sentence:", test_sentence)
    print("Predicted Sentiment:", predicted_sentiment)
    print("Sentiment Score:", predicted_sentiment_score)
```
